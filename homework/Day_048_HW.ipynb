```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
```

```python
train = pd.read_csv('D:/jupyter/ML_100Day_Data/ML_Data/Kaggle_pratice/train.csv', header=None)
test = pd.read_csv('D:/jupyter/ML_100Day_Data/ML_Data/Kaggle_pratice/test.csv', header=None)
train_label = pd.read_csv('D:/jupyter/ML_100Day_Data/ML_Data/Kaggle_pratice/trainLabels.csv', header=None)
```

```python
print(test.shape)
```

```python
GB = GradientBoostingClassifier()
GB.fit(train, train_label)
y_pred = GB.predict(test)
```

```python
Id = np.arange(1,test.shape[0] + 1)
ironmen_dict = {
                "Id": Id,
                "Solution": y_pred
}
submit = pd.DataFrame(ironmen_dict)
print(submit.shape)
submit.to_csv('D:/jupyter/ML_100Day_Data/ML_Data/Kaggle_pratice/submit.csv', index=False)
```
